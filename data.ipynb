```python
import openpyxl
import yfinance as yf
```

```python
wb = openpyxl.load_workbook('NIFTY100_2015.xlsx')

def get_tickers(sheet_name):
    ws = wb[sheet_name]
    return [ws.cell(row=r, column=6).value for r in range(2, ws.max_row + 1)]

tickers_jan = get_tickers('Jan 1 2015')
all_tickers = list(set(tickers_jan))
```

```python
prices = yf.download(all_tickers, start='2015-01-01', end='2016-01-01')['Close']
monthly_prices = prices.resample('ME').last()
monthly_prices.head()
```
